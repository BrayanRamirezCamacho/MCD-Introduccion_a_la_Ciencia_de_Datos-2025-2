# Quién es quién en los precios

La **Procuraduría Federal del Consumidor** (Profeco) es un organismo descentralizado de la Secretaría de Economía del Gobierno de México que tiene como misión empoderar al consumidor mediante la protección efectiva del ejercicio de sus derechos y la confianza ciudadana, promoviendo un consumo razonado, informado, sostenible, seguro y saludable, a fin de corregir injusticias del mercado, fortalecer el mercado interno y el bienestar de la población.

El programa **Quién es Quién en los Precios** recaba y difunde información de precios de productos de consumo regular en el hogar, como *alimentos*, *bebidas*, *productos de aseo personal* y del *hogar*, *medicinas*, *electrodomésticos* y *artículos de temporada* para ofrecer información que te permita tomar decisiones de compra mediante la comparación de precios.

Para recabar los precios se visita una muestra de los principales establecimientos en cada una de las ciudades que forman parte del programa durante los cinco días hábiles de la semana.

Los precios que se muestran contienen la fecha en que fueron tomados, sin embargo, se encuentran sujetos a cambios, ya que algunos establecimientos pueden variar sus precios más de una vez por día, por lo que deben ser considerados como precios referencia.

- La información se refiere a los precios de menudeo vigentes en el momento de su levantamiento.
- Los precios que se publican como referencia para las y los consumidores, son de carácter informativo, por lo que la Procuraduría Federal del Consumidor no es responsable por las modificaciones o variaciones que puedan realizarse por los establecimientos comerciales al día de su publicación ni por el uso que se le dé a la información publicada.

La Profeco pone a disposición la base de datos histórica del programa Quién es Quién en los Precios. Los canales de difusión de estos datos son:
- [La revista del consumidor](https://revistadelconsumidor.profeco.gob.mx/)
- [La mañanera del pueblo](https://www.gob.mx/profeco/archivo/multimedia?&filter_origin=archive)
- [Un histórico de las diapositivas de reportes en formato PDF](https://qqph.profeco.gob.mx/)

Sin embargo, el medio de difusión que más nos interesa a nosotros como científicos de datos en formación es la publicación de [Datos abiertos de este programa](https://historico.datos.gob.mx/busca/dataset/quien-es-quien-en-los-precios), el cuál consiste de un registro histórico diario de más de 2,000 productos, a partir de 2015, en diversos establecimientos de la República Mexicana.

Nos interesa en particular la distribución de datos en formato CSV (Comma-Separated Values).

````python
#Guarda la URL donde están los datos abiertos de PROFEC0.
url = "https://datos.profeco.gob.mx/datos_abiertos/qqp.php#"

#Inicializa soup como None para luego asignarlo cuando se cargue el HTML.
soup = None

try:
    headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept": "text/html,application/xhtml+xml,application/xml",
    "Accept-Language": "es-ES,es;q=0.9",
    "Referer": "https://datos.profeco.gob.mx/"}

    r = req.get(url, headers=headers, timeout=15)
    #Hace una petición HTTP GET al sitio web y devuelve una respuesta (Response object) asignada a r.
    # r = req.get(url)
    
    #Lanza una excepción si el servidor responde con error
    r.raise_for_status()
    #Convierte el HTML crudo en un objeto navegable, a partir del cual se pueden usar selectores CSS para buscar etiquetas.
    soup = BeautifulSoup(r.text)

#Captura errores de red, URLs caídas, bloqueos, y timeouts.
except req.RequestException as e:
    print(f"Error abriendo sitio web: {e}")
````

## Descarga de los archivos desde el 2015 hasta el 2025

````python
#carpeta destino
os.makedirs("QQP_Historico", exist_ok=True)

for link in links:
    #texto del link
    texto = link.text.strip()

    #año desde el texto
    year = texto[-4:]

    #url real usando tu base + href
    url_descarga = url_base + link["href"]

    #nombre local
    nombre = f"QQP_{year}.rar"
    ruta = os.path.join("QQP_Historico", nombre)

    # 🔹 VERIFICAR SI YA EXISTE
    if os.path.exists(ruta) and os.path.getsize(ruta) > 5_000_000:
        print(f"⏭️  {nombre} ya existe — se omite descarga.\n")
        continue

    #descarga
    print(f"Descargando {nombre} ...")
    r = req.get(url_descarga)
    r.raise_for_status()

    #guardado
    with open(ruta, "wb") as f:
        f.write(r.content)

    print(f"✅ Guardado: {ruta}\n")

print("🎯 Descarga completa del histórico.")
````

````python
import subprocess
````

Se descomprime todo en la carpeta de archivos:

````python
carpeta_rar = "QQP_Historico"
carpeta_csv = "QQP_Extraidos"
os.makedirs(carpeta_csv, exist_ok=True)

for archivo in os.listdir(carpeta_rar):
    if archivo.endswith(".rar"):
        ruta = os.path.join(carpeta_rar, archivo)

        # nombre base sin .rar → ej: QQP_2024
        nombre_base = archivo.replace(".rar", "")

        # carpeta donde debería quedar extraído
        carpeta_destino = os.path.join(carpeta_csv, nombre_base)

        # ✅ Si ya existe → no descomprimir otra vez
        if os.path.exists(carpeta_destino):
            print(f"⏭️  {archivo} ya fue descomprimido, se omite.\n")
            continue

        # crear carpeta destino
        os.makedirs(carpeta_destino, exist_ok=True)

        print(f"Extrayendo {archivo} ...")

        resultado = subprocess.run([
            "unrar", "x",
            "-o+",      # sobrescribe sin preguntar
            ruta,
            carpeta_destino
        ])

        # Log de fallo, pero continúa
        if resultado.returncode != 0:
            print(f"⚠️ ERROR extrayendo {archivo} (continuando proceso)\n")
        else:
            print(f"✅ Extraído correctamente: {archivo}\n")
````

# Carga de todos los CSV y filtrado por estado

Ahora, ya que sabemos que el procedimiento funciona, vamos a realizarlo para cargar todos los CSVs. Vamos a cargar cada archivo CSV uno por uno, filtrando para tomar solo los datos del Estado de México y solo entonces unir los dataframes en uno solo (en una celda posterior). Esto es así para no cargar demasiado a la memoria RAM y así evitar que colapse el Kernel (que ya ha pasado en este análisis con enfoques utilizados anteriormente, como unir todos los dataframes y después hacer el filtrado).

````python
dataframes = []
total_archivos = 0
usados = 0

for archivo in csvs:
    total_archivos += 1
    
    try:
        df = pd.read_csv(
            archivo,
            encoding="latin1",
            sep=None,
            engine="python",
            header=None,
            on_bad_lines="skip",
            #low_memory=False
        )

        # asignar columnas
        df.columns = [
            "CLAVE", "PRODUCTO", "MARCA", "CATEGORIA", "TIPO",
            "PRECIO", "FECHA", "CADENA", "GIRO", "SUCURSAL",
            "DIRECCION", "ESTADO", "MUNICIPIO", "LAT", "LON"
        ]

        # corregir encoding del estado
        df["ESTADO"] = df["ESTADO"].astype(str).str.encode("latin1").str.decode("utf-8")

        # filtrar solo EDOMEX
        df = df[df["ESTADO"] == "ESTADO DE MÉXICO"]

        # guardar solo si hay datos útiles
        if len(df) > 0:
            dataframes.append(df)
            usados += 1

    except Exception as e:
        print(f"⚠️ Error en {archivo}: {e}")

print(f"✅ Archivos leídos: {total_archivos}")
print(f"✅ Archivos con datos de EDOMEX: {usados}")
print(f"✅ DataFrames guardados: {len(dataframes)}")
````

Ahora se 'limpian' los productos: se pasa todo a minúsculas; se eliminan acentos; se eliminan caracteres raros; se quitan unidades o formatos como “BOTELLA”, “CAJA”, “LATA”, etc; y se conservan solo letras y espacios.

````python
import unicodedata
import re

def normalizar_producto(p):
    if pd.isna(p):
        return None
    
    # pasar a minúsculas
    p = p.lower()
    
    # eliminar acentos usando unicodedata (estándar)
    p = ''.join(
        c for c in unicodedata.normalize('NFD', p)
        if unicodedata.category(c) != 'Mn'
    )
    
    # reemplazar cualquier cosa que no sea letra/número/espacio
    p = re.sub(r"[^a-z0-9 ]", " ", p)
    
    # colapsar espacios múltiples
    p = re.sub(r"\s+", " ", p).strip()
    
    return p

df_edomex["producto_normalizado"] = df_edomex["PRODUCTO"].apply(normalizar_producto)
````

Con el catálogo de EDOMEX creado, estamos listos para hacer fuzzy matching con la librería RapidFuzz.

````python
from rapidfuzz import process, fuzz

equivalencias = {}

# para cada producto de la canasta básica
for prod_profeco in canasta_df["producto_normalizado"]:
    # se obtiene el producto del catálogo de EDOMEX más parecido en nombre
    mejor_match = process.extractOne(
        query=prod_profeco,
        choices=catalogo_edomex,
        scorer=fuzz.WRatio,
    )

    # si se encontró una equivalencia
    if mejor_match is not None:
        # se toma junto con su puntaje de equivalencia:
        equivalente, score, _ = mejor_match
        # asignarlas como valores al diccionario
        equivalencias[prod_profeco] = (equivalente, score)

equivalencias
````

La celda anterior hace un mapeo de equivalencias entre la canasta básica de PROFECO y el catálogo de productos únicos que hay en el Estado de México. `equivalencias` es un diccionario con pares de producto de la canasta (string) y una dupla consistente en un producto equivalente del catálogo y su correspondiente puntaje de equivalencia (entre más alto el puntaje, más parecido es el producto equivalente al producto de la canasta).

Para calcular la distancia al punto objetivo (la casa de mi compadre, con latitud y longitud = (19.64018, -99.01809)), se utiliza la distancia Manhattan ya que es más cercana a la distancia real que mi compadre tiene que recorrer para llegar a un supermercado en el Estado de México. 

La fórmula en coordenadas (lat, lon) de la distancia Manhattan es:

$𝑑_{Manhattan}=|lat_1 - lat_2| + |lon_1 - lon_2|$

Pero teninendo en cuenta que las coordenadas lat/lon son grados, no distancias físicas, se necesita usar un factor de conversión para convertirlas a kilómetros. Las aproximaciones que se usarán como factores de conversión son: $1°$ $de$ $latitud$ $≈$ $111.32$ $km$ y  $1°$ $de$ $longitud$ $≈$ $111.32$ $km$ $*$ $cos(latitud$ $promedio)$.

Para el punto objetivo ($latitud$ $promedio ≈ 19.64°$):

````python
# constantes de conversión
km_por_grado_lat = 111.32
km_por_grado_lon = 111.32 * np.cos(np.radians(19.64))
print(km_por_grado_lon, " km")
````

````python
import numpy as np

# se define la distancia Manhattan:
def manhattan_km(lat1, lon1, lat2, lon2):
    dlat = abs(lat1 - lat2) * km_por_grado_lat
    dlon = abs(lon1 - lon2) * km_por_grado_lon
    return dlat + dlon
````

Hay que recordar que el objetivo de este análisis era contestar a la pregunta: “¿en qué lugares cerca de tu residencia (o la de mi compadre) se puede comprar una canasta básica a menor precio, tomando en cuenta varios años del histórico de precios?”.

Usando el histórico de precios de PROFECO de la encuesta **"Quién es Quién en los Precios?"** ("QQP") para el Estado de México entre 2015 y 2024, calculé el precio promedio de la canasta básica oficial de PROFECO (24 productos) para cada sucursal de supermercado, considerando únicamente aquellas sucursales donde se puede obtener la canasta  básica completa.

Definí un entorno de “cercanía” como las sucursales ubicadas a menos de 5 km en distancia Manhattan (distancia caminada por las calles, no "a vuelo de pájaro") de la casa de mi compadre en Ecatepec (19.64 N, −99.02 O). Solo 5 km para poder ir a hacer las compras caminando sin tardar demasiado en el traslado.

Para cada sucursal y cada producto de la canasta, hay que calcular el precio promedio histórico:

````python
# Partimos de df_filtrado, que ya sólo tiene productos de la canasta
precios_medios = (
    df_filtrado
    .groupby(["SUCURSAL", "LAT", "LON", "producto_cb"])["PRECIO"]
    .mean()
    .reset_index()
)
````

Después hay que sumar esos precios medios por sucursal:

````python
total_canasta_hist = (
    precios_medios
    .groupby(["SUCURSAL", "LAT", "LON"])["PRECIO"]
    .sum()
    .reset_index()
    .rename(columns={"PRECIO": "total_canasta_hist"})
)

df_completo_hist = total_canasta_hist.copy()
````

Obteniendo así el precio total de la canasta básica para cada sucursal, promediado a lo largo de los años.

````python
df_completo_hist["distancia_manhattan_km"] = df_completo_hist.apply(
    lambda r: dist_manhattan(r["LAT"], r["LON"], lat0, lon0),
    axis=1
)

print(df_completo_hist.shape)
df_completo_hist.head()
````

Se crea un dataframe con las tiendas/sucursales que están a menos de 5 km y se ordenan en base al precio de la canasta básica, de menor a mayor. Con estos candidatos se elegirán las mejores opciones para hacer las compras:

````python
radio_km = 5

candidatas = df_completo_hist[
    df_completo_hist["distancia_manhattan_km"] <= radio_km
].copy()

candidatas_ordenadas = candidatas.sort_values("total_canasta_hist").reset_index()

candidatas_ordenadas[[
    "SUCURSAL", "total_canasta_hist", "distancia_manhattan_km", "LAT", "LON"
]].head(10)
````
